```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
#%config InlineBackend.figure_format = 'svg'
mpl.rcParams['figure.figsize'] = (8.0, 4.0)
mpl.rcParams['figure.dpi'] = 200
```

```python
def likelihood(g, k, l, e=0.001, m=2):
    if g not in [0,1,2]:
        raise ValueError("g must be 0, 1 or 2; %r is not valid" % g)
    if l > k:
         raise ValueError("l cannot be larger thank k")
            
    return (
    (1 / (m**k))
    * (( ((m - g) * e) + (g * (1 - e))) ** l)
    * ((((m - g) * (1 - e)) + (g * e)) ** (k - l))
    )
```

```python
k = 30

ll = np.vstack([np.array([likelihood(g=0, k=k, l=x) for x in range(k+1)]), 
                np.array([likelihood(g=1, k=k, l=x) for x in range(k+1)]),
                np.array([likelihood(g=2, k=k, l=x) for x in range(k+1)])])

for g in ll:
    plt.plot(range(k+1), g, "--o")
```

```python
for g in ll:
    plt.plot(range(k+1)[3:-3], g[3:-3], "--o")
```
